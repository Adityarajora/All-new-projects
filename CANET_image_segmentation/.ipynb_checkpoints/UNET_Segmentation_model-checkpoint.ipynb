```python
def return_file_names_df(root_dir):
    # write the code that will create a dataframe with two columns ['images', 'json']
    # the column 'image' will have path to images
    # the column 'json' will have path to json files
    all_images = []
    all_masks = []
    root = root_dir
    for_image = f"{root}/images"
    for_mask = f"{root}/mask"
    for image_no in os.listdir(for_image):
        for img in os.listdir(f"{for_image}/{image_no}"):
            all_images.append(f"{for_image}/{image_no}/{img}")

    for mask_no in os.listdir(for_mask):
        for msk in os.listdir(f"{for_mask}/{mask_no}"):
            all_masks.append(f"{for_mask}/{mask_no}/{msk}")
    
    data_df = pd.DataFrame(zip(all_images, all_masks), columns=["image", "json"])
    return data_df
```

#### Compute the unique labels

Let's see how many unique objects are there in the json file.
to see how to get the object from the json file please check <a href='https://www.geeksforgeeks.org/read-json-file-using-python/'>this blog </a>

```python
def return_unique_labels(data_df):
    # for each file in the column json
    #       read and store all the objects present in that file
    # compute the unique objects and retrun them
    # if open any json file using any editor you will get better sense of it
    
    unique_labels = set()
    for json_file in tqdm(data_df['json'].values):
        f = open(json_file)
        data = json.load(f)

        
        for obj in data['objects']: unique_labels.add(obj['label'])
            
    return unique_labels
```

## 3. Extracting the polygons from the json files

```python
def get_poly(file):
    # this function will take a file name as argument
    
    # it will process all the objects in that file and returns
    
    # label: a list of labels for all the objects label[i] will have the corresponding vertices in vertexlist[i]
    # len(label) == number of objects in the image
    
    # vertexlist: it should be list of list of vertices in tuple formate 
    # ex: [[(x11,y11), (x12,y12), (x13,y13) .. (x1n,y1n)]
    #     [(x21,y21), (x22,y12), (x23,y23) .. (x2n,y2n)]
    #      .....
    #     [(xm1,ym1), (xm2,ym2), (xm3,ym3) .. (xmn,ymn)]]
    # len(vertexlist) == number of objects in the image
    
    # * note that label[i] and vertextlist[i] are corresponds to the same object, one represents the type of the object
    # the other represents the location
    
    # width of the image
    # height of the image
    
    f = open(file)
    data = json.load(f)
    
    w, h = data["imgWidth"], data["imgHeight"]
    label, vertexlist = [], []
    for obj in data['objects']:
        label.append(obj["label"])
        vertexlist.append([(x,y) for x, y in obj["polygon"]])
        
    return w, h, label, vertexlist
```

### Computing mask arrays for each image

```python
def compute_masks(data_df):
    # after you have computed the vertexlist plot that polygone in image like this
    
    # img = Image.new("RGB", (w, h))
    # img1 = ImageDraw.Draw(img)
    # img1.polygon(vertexlist[i], fill = label_clr[label[i]])
    
    # after drawing all the polygons that we collected from json file, 
    # you need to store that image in the folder like this "data/output/scene/framenumber_gtFine_polygons.png"
    
    # after saving the image into disk, store the path in a list
    # after storing all the paths, add a column to the data_df['mask'] ex: data_df['mask']= mask_paths
    
    mask_img_path = []
    for json_file in tqdm(data_df["json"]):
        
        w, h, labels, vertexlist = get_poly(json_file)
        
        img_path = json_file.replace("mask", "output").replace("json", "png")
        
        os.makedirs(os.path.dirname(img_path), exist_ok=True)
        
        img = Image.new("RGB", (w, h))
        img1 = ImageDraw.Draw(img)
        
        
        for i in range(len(labels)):
            if len(vertexlist[i]) < 2:
                print(f"No polyline for that object {len(vertexlist[i])}, skipping!->{labels[i]}")
                continue
            img1.polygon(vertexlist[i], fill = label_clr[labels[i]]//10)
   
            
        img=np.array(img)
        im = Image.fromarray(img[:,:,0])
        im.save(img_path)
        
        mask_img_path.append(img_path)
    
    data_df["mask"] = mask_img_path    
    return data_df
```

```python
def visualize(**images):
    n = len(images)
    plt.figure(figsize=(16, 5))
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        if i==1:
            plt.imshow(image, cmap='gray', vmax=1, vmin=0)
        else:
            plt.imshow(image)
    plt.show()
    
def normalize_image(mask):
    mask = mask/255
    return mask

class Dataset:
    # we will be modifying this CLASSES according to your data/problems
    CLASSES = [0,130,10,140,20,150,30,160,40,170,50,180,60,190,70,200,80,90,100,110,120]
    
    # the parameters needs to changed based on your requirements
    # here we are collecting the file_names because in our dataset, both our images and maks will have same file name
    # ex: fil_name.jpg   file_name.mask.jpg
    def __init__(self, img_path):
        
        self.ids = img_path
        self.images_fps = [img for img in img_path]
        # the paths of segmentation images
        self.masks_fps = [img.replace("leftImg8bit.jpg", "gtFine_polygons.png").replace("images", "output") for img in img_path]
        # giving labels for each class

        self.class_values = Dataset.CLASSES
        
        
        
        
    
    def __getitem__(self, i):
        
        # read data
        old_dimn = (1080, 1920)
        new_dimn = (512, 512)
        image = cv2.imread(self.images_fps[i], cv2.IMREAD_UNCHANGED)
        image = cv2.resize(image, new_dimn, interpolation = cv2.INTER_NEAREST)
        image = image/255

        
        mask  = cv2.imread(self.masks_fps[i], cv2.IMREAD_UNCHANGED)
        image_mask = cv2.resize(mask, new_dimn, interpolation = cv2.INTER_NEAREST)
        
    
        
        image_mask = [(image_mask == v//10) for v in self.class_values]
        image_mask = np.stack(image_mask, axis=-1).astype('float')
        
#         print(np.max(image))
        a = np.random.uniform()
        if a<0.2:
            image = aug2.augment_image(image)
            image_mask = aug2.augment_image(image_mask)
        elif a<0.4:
            image = aug3.augment_image(image)
            image_mask = aug3.augment_image(image_mask)
        elif a<0.6:
            image = aug4.augment_image(image)
            image_mask = aug4.augment_image(image_mask)
        elif a<0.8:
            image = aug5.augment_image(image)
            image_mask = image_mask
        else:
            image = aug6.augment_image(image)
            image_mask = aug6.augment_image(image_mask)
#         print(np.max(image_mask), "A")    
        return image, image_mask
        
    def __len__(self):
        return len(self.ids)
    
    
class Dataloder(tf.keras.utils.Sequence):    
    def __init__(self, dataset, batch_size=1, shuffle=False):
        self.dataset = dataset
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.indexes = np.arange(len(dataset))

    def __getitem__(self, i):
        
        # collect batch data
        start = i * self.batch_size
        stop = (i + 1) * self.batch_size
        data = []
        for j in range(start, stop):
            data.append(self.dataset[j])
        
        batch = [np.stack(samples, axis=0) for samples in zip(*data)]
        
        return tuple(batch)
    
    def __len__(self):
        return len(self.indexes) // self.batch_size
    
    def on_epoch_end(self):
        if self.shuffle:
            self.indexes = np.random.permutation(self.indexes)
```

### Task 2.1: Dice loss

<pre>
* Explain the Dice loss
* 1. Write the formualtion
* 2. Range of the loss function
* 3. Interpretation of loss function
* 4. Write your understanding of the loss function, how does it helps in segmentation
</pre>

* Dice Loss basically based on Dice coefficient
* Dice coefficient is the scalar value which represents how much tow sets overlapping
* 0 <= Dice coefficient <= 1
* Dice coefficient = 0 => Two sets has no overlapping
* Dice coefficient = 1 => Two sets are perfectly overlapping
* Dice coefficient for two sets:-
    # DC = 2|A intersection B| / |A| + |B|
* Dice coefficient for image segmentation:-
    A-> matrix actual maks of the image
    B-> matrix predicted mask by a model like UNet
    
    DC = 2 * **sum(element_wise_multiplication(A,B)) /  sum(A) + sum(B)**
    
    - element_wise_multiplication is concerned about the total activations, we are discarding inactive (wiz 0 intensity)
* Dice loss = 1 - Dice coefficient
    
    - So, as Dice coefficient is large means actual and predicted masks are highly overlapping and Loss will be less
    as we are subtracting it from 1
    - If  Dice coefficient is very small means actual and predicted masks are not overlapping to Loss will be large

```python
import tensorflow as tf
# tf.compat.v1.enable_eager_execution()
from tensorflow import keras
from tensorflow.keras.layers import *
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import UpSampling2D
from tensorflow.keras.layers import MaxPooling2D, GlobalAveragePooling2D
from tensorflow.keras.layers import concatenate
from tensorflow.keras.layers import Multiply
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Input, Add, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPooling2D, GlobalMaxPooling2D
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.utils import plot_model
from tensorflow.keras.initializers import glorot_uniform
K.set_image_data_format('channels_last')
K.set_learning_phase(1)
```

 
* You can devide the whole architecture can be devided into two parts
    1. Encoder
    2. Decoder
    <img src='https://i.imgur.com/prH3Mno.png' width="600">
* Encoder:
    * The first step of the encoder is to create the channel maps [$C_1$, $C_2$, $C_3$, $C_4$]
    * $C_1$ width and heigths are 4x times less than the original image
    * $C_2$ width and heigths are 8x times less than the original image
    * $C_3$ width and heigths are 8x times less than the original image
    * $C_4$ width and heigths are 8x times less than the original image
    * <i>you can reduce the dimensions by using stride parameter</i>.
    * [$C_1$, $C_2$, $C_3$, $C_4$] are formed by applying a "conv block" followed by $k$ number of "identity block". i.e the $C_k$ feature map will single "conv block" followed by $k$ number of "identity blocks".
    <table>
    <tr><td><img src="https://i.imgur.com/R8Gdypo.png" width="300"></td>
        <td><img src="https://i.imgur.com/KNunjQK.png" width="250"></td></tr>
    </table>
    * <strong>The conv block and identity block of $C_1$</strong>: the number filters in the covolutional layers will be $[4,4,8]$ and the number of filters in the parallel conv layer will also be $8$.
    * <strong>The conv block and identity block of $C_2$</strong>: the number filters in the covolutional layers will be $[8,8,16]$ and the number of filters in the parallel conv layer will also be $16$.
    * <strong>The conv block and identity block of $C_3$</strong>: the number filters in the covolutional layers will be $[16,16,32]$ and the number of filters in the parallel conv layer will also be $32$.
    * <strong>The conv block and identity block of $C_4$</strong>: the number filters in the covolutional layers will be $[32,32,64]$ and the number of filters in the parallel conv layer will also be $64$.
    * Here $\oplus$ represents the elementwise sum
    <br>
    
    <font color="red">NOTE: these filters are of your choice, you can explore more options also</font>
    
    * Example: if your image is of size $(512, 512, 3)$
        * the output after $C_1$ will be $128*128*8$
        * the output after $C_2$ will be $64*64*16$
        * the output after $C_3$ will be $64*64*32$
        * the output after $C_4$ will be $64*64*64$
